### 머신러닝 워크플로우 (Machine Learning Workflow)
1. 머신러닝 데이터 수집
 - 머신러닝 워크플로우에서 데이터 수집은 매우 중요한 과정 
 - 데이터 수집 중, 수집한 데이터의 품질로 프로젝트의 잠재적 유용성과 정확도를 정의할 수 있음
 - 데이터 수집을 위해, 자원을 식별하고 식별된 자원으로 데이터를 통합하여 단일 데이터셋으로 변경할 필요가 있음
 - 이 데이터셋은 사물인터넷 센서에서 데이터를 스트리밍하거나, 오픈 데이터 소스를 다운로드하거나, 여러 파일, 로그, 또는 미디어로부터 데이터 레이크를 구성하는 것을 의미

2. 데이터 전처리
 - 데이터가 수집되면 전처리가 필요함
 - 전처리 과정은 정제, 검증, 사용가능한 데이터셋으로 형변환 등을 포함
 - 단일 소스 데이터를 수집한 경우 간단한 과정으로 처리가 가능
 - 그러나, 여러 개의 데이터를 수집한 경우, 데이터 형식이 일치 하는지 동일하게 신뢰할 수 있는지 확인해야하고 잠재적 중복 요소를 제거해야 함

3. 데이터셋 구축
 - 이 단계는 데이터를 3개의 데이터셋으로 분할하는 과정이 포함됨 : 훈련, 검증, 테스트
 - Training set : 알고리즘을 처음 훈련시키는데 사용하고 정보처리 방법을 학습. 이 데이터셋은 파라미터를 통한 모델 분류를 정의
 - Validation set : 모델의 정확도에 대해 평가하기 위해 사용. 이 데이터셋은 모델 파라미터를 미세하게 조정하기 위해 사용
 - Test set : 모델의 정확도와 성능에 대해 평가하기 위해 사용. 이 데이터셋은 모델에서 잘못된 훈련이나 특정 이슈를 찾기 위해 사용

4. 훈련 및 정제
 - 알고리즘에 훈련 세트를 제공하여 분류에 사용되는 적절한 매개변수와 특징을 훈련할 수 있음
 - 훈련이 완료되면 유효성 검사 데이터셋을 사용하여 모델을 정제할 수 있음
 - 이 정제과정은 변수 수정 및 삭제가 포함될 수 있고, 허용 가능한 정확도 수준에 도달 할 때까지 모델 별 설정(하이퍼 매개 변수)을 조정하는 프로세스가 포함됨

5. 머신러닝 평가
 - 허용 가능한 하이퍼 파라미터 세트를 찾고 모델 정확도를 최적화 한 후 모델을 테스트 할 수 있음
 - 테스트는 테스트 데이터 세트를 사용하며 모델이 정확한 기능을 사용하고 있는지 확인하기위한 것
 - 확인 후 받은 피드백에 따라 정확도를 높이고 출력 설정을 조정하거나 필요에 따라 모델을 배포하기 위해 모델 학습으로 돌아갈 수 있음

#### 퀴즈
- 크롤링 환경을 구축하여 서울시 게시판에 있는 데이터 중 title 태그의 값을 가져와서 출력하기

```python
# 페이지 번호 1 ~ 10번까지의 기사 제목를 리스트에 저장하기
titleList = []
query = parse.quote("빅데이터")
for i in range(1, 50, 1): # 1. start 번호가 1, 11, 21, ... , 91이 되도록 범위 설정
    start = str(i)
    # 2. 각 start 번호 별로 데이터 추출
    url = "https://news.nate.com/search?q=" + query + "&page=" + start
    html = urlopen(url) 
    # 3. 가져온 데이터 html로 파싱
    data = BeautifulSoup(html, "html.parser")
    # 4. 파싱된 html에서 title 데이터만 추출하여 저장
    for title in data.find_all("h2", {"class":"tit"}):
        titleList.append(title.text)
```

## 텍스트 전처리(Text Preprocessing)
- 텍스트 전처리는 용도에 맞게 텍스트를 사전에 처리하는 과정
- 필요한 텍스트만 용도에 맞게 처리해야 데이터를 사용할 수 있음
- 크롤링 등으로 얻어낸 코퍼스(corpus, 말뭉치) 데이터는 용도에 맞게 전처리 과정을 거쳐야 함
- 이 과정은 토큰화, 정제, 정규화 등을 포함

### 토큰화(Tokenization)
- 주어진 코퍼르에서 토큰이라고 불리는 단위로 나누는 작업을 토큰화라고 함
- 토큰의 단위는 상황에 따라 다르지만, 보통 의미있는 단위로 토큰을 정의

#### 한국어 토큰화 실습
- 형태소(morpheme)란 뜻을 가진 가장 작은 말의 단위를 의미
- 한국어 토큰화에는 형태소 개념에 대한 이해가 필요하며, 형태소에는 자립 형태소와 의존 형태소가 존재
- 자립 형태소 : 접사, 어미, 조사와 상관없이 자립하여 사용할 수 있는 형태소. 체언(명사, 대명사, 수사), 수식언(관형사, 부사), 감탄사 등이 있음
- 의존 형태소 : 다른 형태소와 결합하여 사용되는 형태소. 접사, 어미, 조사, 어간 등이 있음
- 예시) 둘리가 집을 나갔다.
 - 자립 형태소 : 둘리, 집
 - 의존 형태소 : -가, -을, 나-, -갔, -다
- 형태소 토큰화를 통해 의미있는 단어를 추출
- 한국어 자연어 처리는 KoNLPy(코엔엘파이) 패키지를 사용
- 코엔엘파이 형태소 분석기는 Okt(Open Korea Text), 메캅(Mecab), 코모란(Komoran), 한나눔(Hannanum), 꼬꼬마(Kkma)가 있음

```python
from konlpy.tag import Okt # Open Korea Text 라이브러리 임포트   
okt=Okt()  
# morphs : 형태소 추출
# pos : 품사 태깅(Par-of-speech tagging)
# nouns : 명사 추출
print(okt.morphs("안녕하세요. 날씨가 좋네요. 다같이 즐겁게 코딩합시다!!"))
```

```python
print(okt.pos("안녕하세요. 날씨가 좋네요. 다같이 즐겁게 코딩합시다!!"))
```

```python
print(okt.nouns("안녕하세요. 날씨가 좋네요. 다같이 즐겁게 코딩합시다!!"))
```

```python
# 네이버 빅데이터 검색 결과 데이터에서 명사만 추출하기
titleNounsList = []
for title in titleList:
    list = okt.nouns(title)
    for l in list:
        titleNounsList.append(l)
        print(l)
        
```

### 정제(Cleaning)와 정규화(Normalization)
- 정제 : 추출된 코퍼스로부터 결측값 및 필요없는 데이터를 제거
- 정규화 : 같은 의미나 중복된 단어들을 통합
- 정제 작업의 경우 한 번만 수행하는 것이 아니라 토큰화 작업 후에 남아있는 데이터들을 제거하기위해 반복수행할 수 있음

#### 불필요한 단어의 제거
- 불필요한 단어는 아무 의미가 없는 글자나 단어, 혹은 분석 목적과 맞지 않는 단어를 의미
- 또한, 단어의 등장 빈도가 적다면 자연어 처리에 특별하게 도움되지 않기 때문에 제거하는 것이 좋음
- 단어의 길이가 짧은 단어도 의미가 없을 수 있음. 예를 들어, 차, 군, 인 등 의미가 중복되거나 단어 자체만으로는 의미가 모호한 단어들은 제거할 필요가 있음

```python
# 네이버 뉴스 검색 결과에서 한 글자의 단어 제거
result = []
for l in titleNounsList:
    if len(l) > 1 :
        result.append(l)

print(result)
```

```python
from collections import Counter

Counter(result) # 입력된 데이터에서 중복 데이터에 대한 빈도수를 표현
Counter(result).most_common() # 중복 데이터에 대한 빈도수를 내림차순으로 정렬하여 튜플로 표현
```

```python
# 글자의 빈도수가 25이하인 데이터는 제외
freq_result = []
for data, c in Counter(result).most_common():
    if c > 25 :
        freq_result.append((data, c))
print(freq_result)
```

#### 데이터 시각화
- 자연어 처리를 위한 텍스트 분석에는 딥 러닝에 기반한 분석 방식을 사용해야함
- 딥러닝 분석의 기본적인 구조는 언어 패턴을 분석하여 단어와 뜻을 기계에 학습시키는 방식
- 이는 통계적 구조 및 수학적 수식에 대한 이해가 필요하므로 이 과정에서는 다루지 않음
- 간단하게 빈도수를 그림으로 나타내는 빈도 막대그래프와 워드 클라우드 라이브러리를 이용해 시각적으로 표현할 예정

```python
dataList = []
indexList = []
for data, index in freq_result:
    dataList.append(data)
    indexList.append(index)

# 그래프를 노트북 안에 그리기 위해 설정
%matplotlib inline

# 필요한 패키지와 라이브러리를 가져옴
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
plt.rcParams['axes.unicode_minus'] = False
print ('버전: ', mpl.__version__)
print ('설치 위치: ', mpl.__file__)
print ('설정 위치: ', mpl.get_configdir())
print ('캐시 위치: ', mpl.get_cachedir())
print ('설정 파일 위치: ', mpl.matplotlib_fname())

# 나눔 폰트를 설정하기 전에 캐시 위치에 있는 파일을 삭제할 것
plt.rcParams['font.family'] = "NanumGothic"
print('# 설정되어있는 폰트 사이즈')
print (plt.rcParams['font.size'] ) 
print('# 설정되어있는 폰트 글꼴')
print (plt.rcParams['font.family'] )

plt.title('빅데이터 검색 단어 수')
plt.barh(dataList, indexList)
```
